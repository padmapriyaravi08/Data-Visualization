```python
#Traversal
arr=[10,20,30,40,50]
for element in arr:
    print(element)
```

```python
#insertion
arr=[1,2,3,4,5,6]
arr.insert(2,60)
print(arr)
```

```python
#Deletion
arr=[10,20,30,40,50,60]
arr.remove(40)
print(arr)
```

```python
#search
arr=[10,20,30,40,50,60]
key=30
if key in arr:
    print("Element Found")
else:
    print("Element not Found")
```

```python
arr = [10, 20, 30, 40, 50, 60]
key = 40

for i in arr:
    if i == key:
        print("Element found")
        break
else:
    print("Element Not Found")

```

```python
#Sort
arr=[5,2,7,3,4,6]
arr.sort()
print(arr)
```

```python
#Update
arr=[1,2,3,4,5,6]
arr[2]=60
print(arr)
```

```python
#Access
arr=[1,2,3,4,5,6]
arr[3]
```
