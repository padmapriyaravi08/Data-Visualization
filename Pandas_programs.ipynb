```python
#create a DataFrame containing rollno,name,department
import pandas as pd
data={'RollNo':[101,102,103,104,105,106,107,108,109,110],
     'Name':['Arun','Priya','Rahul','padma','ravi','rohith','easwari','sanjeeve','mahilini','rose'],
     'Department':['CS','AI','ML','DS','BSC','MBA','B.com','MSc','B.tech','BBA'],
     'Marks':[85,90,78,99,96,96,88,98,86,96]}
df=pd.DataFrame(data)
print(df)
```

```python
#create a dataframe and find total,average,highest and lowest marks
print("Total_Marks=",df['Marks'].sum())
print("Average_Marks=",df['Marks'].mean())
print("Highest_Marks=",df['Marks'].max())
print("Lowest_Marks=",df['Marks'].min())
```
